```python
import pandas as pd
import matplotlib.pyplot as plt

model = pd.read_csv("population_by_age_group.csv")

print(model.head())
```

```python
#Bar graph of Total population against Country
x_value = model["Country"]
y_value = model["Total"]

plt.xlabel("Country")
plt.ylabel("Total population")

plt.yscale('log')
plt.xticks(rotation=45)

plt.bar(x_value, y_value)
plt.show()
```

```python
#Bar graph of Population of 0-4 years against Country
x_value = model["Country"]
y_value = model["0-4 years"]

plt.xlabel("Country")
plt.ylabel("Population of 0-4 years")

plt.yscale('log')
plt.xticks(rotation=45)

plt.bar(x_value, y_value, color='orange')
plt.show()
```

```python
#Bar graph of Population of 65+ against Country
x_value = model["Country"]
y_value = model["65+"]

plt.xlabel("Country")
plt.ylabel("Population of 65+")

plt.yscale('log')
plt.xticks(rotation=45)

plt.bar(x_value, y_value, color='grey')
plt.show()
```
